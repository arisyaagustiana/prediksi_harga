```python
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
import sklearn.metrics as metric
%matplotlib inline
```

```python
filedata='/Users/aguswahyu/Desktop/iris.csv'
data1=pd.read_csv(filedata)
data1
```

```python
X1=data1['sepal-length']
X2=data1['sepal-width']
X_training=np.array(list(zip(X1,X2)))
```

```python
y_training=data1['class']
y_training
```

```python
target_names=['Iris-sentosa', 'Iris-versicolor']
target_names
```

```python
class1=y_training[y_training=='Iris-setosa'].index
class2=y_training[y_training=='Iris-versicolor'].index

plt.scatter(X_training[class1,0],X_training[class1,1],c='b',s=50)
plt.scatter(X_training[class2,0],X_training[class2,1],c='r',s=50)
plt.xlabel("sepal-length")
plt.ylabel("sepal-width")
plt.legend(target_names,loc=2)
plt.savefig('chart0.png')
```

```python
svc = svm.SVC(kernel="poly", gamma="auto").fit(X_training,y_training)
svc
```

```python
lbX1=math.floor(min(X_training[:,0]))-1
ubX1=math.ceil(max(X_training[:,0]))+1
lbX2=math.floor(min(X_training[:,1]))-1
ubX2=math.ceil(max(X_training[:,1]))+1
[lbX1,ubX1,lbX2,ubX2]
```

```python
class1=y_training[y_training=='Iris-setosa'].index
class2=y_training[y_training=='Iris-versicolor'].index
    
plt.scatter(X_training[class1,0],X_training[class1,1],c='b',s=50)
plt.scatter(X_training[class2,0],X_training[class2,1],c='r',s=50)
plt.xlabel("sepal-length")
plt.ylabel("sepal-width")
plt.legend(target_names,loc=2)


X,Y = np.mgrid[lbX1:ubX1:100j,lbX2:ubX2:100j]
Z = svc.decision_function(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
plt.contourf(X,Y,Z > 0,alpha=0.4)
plt.contour(X,Y,Z,colors=['k'], linestyles=['-'],levels=[0])

plt.title('Polynomial Separable')
plt.savefig('chart1.png')
```

```python
class1=y_training[y_training=='Iris-setosa'].index
class2=y_training[y_training=='Iris-versicolor'].index
    
plt.scatter(X_training[class1,0],X_training[class1,1],c='b',s=50)
plt.scatter(X_training[class2,0],X_training[class2,1],c='r',s=50)
plt.xlabel("sepal-length")
plt.ylabel("sepal-width")

plt.legend(target_names,loc=2)
X,Y = np.mgrid[lbX1:ubX1:100j,lbX2:ubX2:100j]
Z = svc.decision_function(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)

plt.contourf(X,Y,Z > 0,alpha=0.4)
plt.contour(X,Y,Z,colors=['k','k','k'], linestyles=['--','-','--'],levels=[-1,0,1])
plt.scatter(svc.support_vectors_[:,0],svc.support_vectors_[:,1],s=120,facecolors='none')

color = []
for i in y_training:
    if i == 'Iris-setosa':
        color.append(1)
    else:
        color.append(-1)
        

plt.scatter(X_training[:,0],X_training[:,1],c=color,s=50,alpha=0.65);

plt.title('Margin and Support Vectors')
plt.savefig('chart2.png')
```

```python
filedata='/Users/aguswahyu/Desktop/iris-test.csv'
data2=pd.read_csv(filedata)

X1=data2['sepal-length']
X2=data2['sepal-width']

X_testing=np.array(list(zip(X1,X2)))
y_testing = data2['class']
```

```python
predict = svc.predict(X_testing)
from sklearn.metrics import accuracy_score
accuracy_score(y_testing, predict)
```
